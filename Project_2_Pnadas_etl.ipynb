### Transform Deflator DataFrame

```python
# Create a filtered dataframe from specific columns
Deflator_cols = ["License Serial Number", "Premises Name", "County ID Code"]
Deflator_transformed= Deflator_df[Deflator_cols].copy()

# Rename the column headers
Deflator_transformed = Deflator_transformed.rename(columns={"License Serial Number": "id",
                                                          "Premises Name": "premise_name",
                                                          "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
Deflator_transformed.drop_duplicates("id", inplace=True)
Deflator_transformed.set_index("id", inplace=True)

Deflator_transformed.head()
```
